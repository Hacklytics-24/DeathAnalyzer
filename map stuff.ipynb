```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python

# Load the dataset
df = pd.read_csv("cause_of_deaths.csv", delimiter=",")

# Calculate total deaths for each country
df['Total_Deaths'] = df.iloc[:, 3:].sum(axis=1)

# Create a choropleth map using Plotly Express
fig = px.choropleth(df, 
                    locations="Country/Territory", 
                    locationmode='country names',
                    color="Total_Deaths",
                    hover_name="Country/Territory",
                    hover_data=["Total_Deaths"],
                    title="Health Concerns by Country",
                    labels={'Total_Deaths':'Total Deaths'},
                    color_continuous_scale=px.colors.sequential.Plasma)

# Show the map
fig.show()
```
